```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained('facebook/bart-large-cnn')
```

```python
targets = ["I have an apple.", "I have a banana."]
with tokenizer.as_target_tokenizer():
    # labels = tokenizer(targets, max_length=100, truncation=True)
    labels = tokenizer(targets)
labels['input_ids']
```

```python
with open("../data/cnn_cln/test.target", "r") as f:
    targets = f.readlines()
targets[:5]
```

```python
with tokenizer.as_target_tokenizer():
    labels = tokenizer(targets, max_length=220, truncation=True)
labels['input_ids']
```

```python
target_len_list = [len(token_ids) for token_ids in labels['input_ids']]
```

```python
with open("../data/cnn_cln/train.target", "r") as f:
    train_targets = f.readlines()
train_targets[:3]
```

```python
with tokenizer.as_target_tokenizer():
    labels = tokenizer(train_targets, max_length=220, truncation=True)
train_target_len_list = [len(token_ids) for token_ids in labels['input_ids']]
labels['input_ids'][:3]
```

```python
plt.hist(train_target_len_list, bins=32)
plt.show()
```

```python
print((np.array(train_target_len_list)>=90).sum())
print(((np.array(train_target_len_list)<90)*(np.array(train_target_len_list)>=70)).sum())
print(((np.array(train_target_len_list)<70)*(np.array(train_target_len_list)>=50)).sum())
print((np.array(train_target_len_list)<50).sum())
```

```python
with open("../data/cnn_cln/val.target", "r") as f:
    val_targets = f.readlines()
with tokenizer.as_target_tokenizer():
    labels = tokenizer(val_targets, max_length=100, truncation=True)
val_target_len_list = [len(token_ids) for token_ids in labels['input_ids']]
plt.hist(val_target_len_list, bins=32)
plt.show()
```

```python
print((np.array(val_target_len_list)>=90).sum())
print(((np.array(val_target_len_list)<90)*(np.array(val_target_len_list)>=70)).sum())
print(((np.array(val_target_len_list)<70)*(np.array(val_target_len_list)>=50)).sum())
print((np.array(val_target_len_list)<50).sum())
```
